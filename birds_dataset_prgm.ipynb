```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data_csv=pd.read_csv("data_csv.csv")
plt.figure(figsize=(10,7))
x=data_csv["Value"]
plt.hist(x,bins=30,color="green")
plt.xlabel("Value")
plt.ylabel("year")
plt.show
```
